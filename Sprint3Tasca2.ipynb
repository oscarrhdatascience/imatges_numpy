- Exercici 4

Mostreu-me amb exemples de diferents matrius, la regla fonamental de Broadcasting que diu : "les matrius es poden transmetre / broadcast si les seves dimensions coincideixen o si una de les matrius té una mida d'1".

```python
# Broadcasting 2 arrays with the same dimensions
arr1 = np.random.randint(1,100, size=(3, 5))
print (arr1)
arr2 = np.random.randint(1,100, size=(3, 5))
print (arr2)
print (arr1+arr2)
print (arr2-arr1)
print (arr1*arr2)
print (arr2/arr1)
```

```python
# Broadcasting an array with 2 dimensions and another with 1 dimension
arr1D=np.random.randint(1,100, size=(5))
print (arr1D)
print (arr1+arr1D)
print (arr1D-arr1)
print (arr1*arr1D)
print (arr1D/arr1)
```

```python
# Broadcasting arrays with different dimensions ----> This won't work
arr3 = np.random.randint(1,100, size=(3, 5, 3))
print (arr3)
print (arr1+arr3)
```

```python
# Flattening an array so we can still broadcasting
arr4=np.array([[[6,]]])
print (arr4)
arr4=arr4.reshape(-1)
print (arr4)
print (arr2+arr4)
print (arr4-arr2)
print (arr2*arr4)
print (arr4/arr2)
```

```python
# Reshaping the arr5 array with the same dimensions of arr2 array so we can still broadcasting
arr5 = np.random.randint(1,100, size=(5, 3))
print (arr5)
arr5 = arr5.reshape(3,5)
print (arr5)
print (arr5+arr2)
print (arr2-arr5)
print (arr5*arr2)
print (arr5/arr2)
```
